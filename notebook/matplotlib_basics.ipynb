```python
# plt.style.context use style settings temporarily for a block of code without changing the global styling
data=np.random.rand(20)
plt.style.context('seaborn-whitegrid')
# marker :'ro' for red circles
plt.plot(data,'ro')
plt.show()
```

```python
# plt.style.context use style settings temporarily for a block of code without changing the global styling
data=np.random.rand(20)
plt.style.context('ggplot')
# marker :'ro' for red circles
plt.plot(data,'ro')
plt.show()
```

```python
# plt.style.context use style settings temporarily for a block of code without to changing the global styling
data=np.random.rand(20)
plt.style.context('fivethirtyeight')
# marker :'ro' for blue circles with lines connecting the circle
plt.plot(data,'b-o')
plt.show()
```

```python
# plt.style.context use style settings temporarily for a block of code without to changing the global styling
data=np.random.rand(20)
plt.style.context('seaborn-notebook')
# marker :'ro' for red circles
plt.plot(data,"r--+")
plt.show()
```

```python
# plt.style.context use style settings temporarily for a block of code without to changing the global styling
data=np.random.rand(20)
plt.style.context('seaborn-notebook')
# marker :'ro' for red circles
plt.plot(data,"r+",linewidth=10)
plt.show()
```

```python
# plt.style.context use style settings temporarily for a block of code without to changing the global styling
data=np.random.rand(20)
plt.style.context('seaborn-notebook')
# marker :'g' for green 
plt.plot(data,"g+--",linewidth=2,markersize=15)
plt.show()
```

```python
# plt.style.context use style settings temporarily for a block of code without to changing the global styling
data=np.random.rand(20)
with plt.style.context('dark_background'):
    # marker :'ro' for red circles
    # linestyle or ls
    plt.plot(data,'ro',ls='solid',linewidth=2,markersize=10,markerfacecolor='green')
plt.show()
```
